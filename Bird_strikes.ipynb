```python
columns_to_fill = ['NumberStruckActual', 'Altitude', 'PeopleInjured']
for col in columns_to_fill:
    df[col] = df[col].fillna(df[col].mean())
```

```python
# bar diagram for altitude type accident
a_count = df['AltitudeBin'].value_counts()

# plot the bar diagram

plt.figure(figsize=(10,6))
a_count.plot(kind='bar', color = 'skyblue')
plt.title('Frequency of Bird Strikes by Altitude')
plt.xlabel('Altitude')
plt.ylabel('Number of Bird Strikes')
plt.xticks(rotation = 45, ha='right')
plt.tight_layout()
plt.show()
```

<!-- The highest number of strikes occurred in (-1,0) altitude below sea level that is on the runway itself.-->
<!-- Lowest is on (5000,18000) altitude from sea level. -->

```python
# convert 'FlightDate' column to datetime format
df['FlightDate'] = pd.to_datetime(df['FlightDate'], errors = 'coerce')

# extract year,month, and day
df['Year'] = df['FlightDate'].dt.year
df['Month'] = df['FlightDate'].dt.month
df['Day'] = df['FlightDate'].dt.day

# display updated data
df[['FlightDate','Year','Month','Day']].head()

```

```python
# group bird strike by month
monthly_strikes = df.groupby('Month').size()
print(monthly_strikes)

yearly_strikes = df.groupby('Year').size()
print(yearly_strikes)
```

```python
# by month
plt.figure(figsize=(10,6))
monthly_strikes.plot(kind='bar', color = 'orange', edgecolor = 'black')

# add titles and labels
plt.title('Bird Strikes by Month', fontsize=16)
plt.xlabel('Month', fontsize = 14)
plt.ylabel('Number of Bird Strikes', fontsize = 14)
plt.xticks(ticks=range(12), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'], rotation=45)
plt.tight_layout()

# Show the plot
plt.show()


# by day
plt.figure(figsize=(10,6))
yearly_strikes.plot(kind='bar', color = 'orange', edgecolor = 'black')

# add titles and labels
plt.title('Bird Strikes by Year', fontsize=16)
plt.xlabel('Year', fontsize = 14)
plt.ylabel('Number of Bird Strikes', fontsize = 14)
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()
```

<!-- It looks like the bird strikes are high in the month of August and it is also increasing yearly. -->

```python
# group by bird species and count of occurrences
top_birds = df['WildlifeSpecies'].value_counts().head(5)
print(top_birds)

# filter data for only top 5 species
top_birds_data = df[df['WildlifeSpecies'].isin(top_birds.index)]
```

```python

# Group by bird species and calculate average altitude
bird_altitudes = top_birds_data.groupby('WildlifeSpecies')['Altitude'].mean()

# Display the average altitudes
print(bird_altitudes)
```

```python
# Plot a bar chart
plt.figure(figsize=(10, 6))
bird_altitudes.sort_values().plot(kind='bar', color='skyblue', edgecolor='black')

# Add titles and labels
plt.title('Average Altitudes of Bird Strikes by Top 5 Bird Species', fontsize=16)
plt.xlabel('Bird Species', fontsize=14)
plt.ylabel('Average Altitude (Feet)', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# 1. Altitude and Number of Strikes (Bar Diagram)
# Insights:

# Altitudes below 3,000 feet likely experience the highest frequency of bird strikes. This is expected as most strikes occur during takeoff and landing.
# Strikes significantly decrease at higher altitudes, but there might still be notable activity up to 10,000 feet, especially involving migratory birds.
# Implications:

# Focus preventive measures, such as bird deterrent systems, during low-altitude flight phases like takeoff and landing.
# Increase radar surveillance or bird tracking up to 10,000 feet in migration zones.
# 2. Strikes by Month and Year (Bar Diagram)
# Insights:

# Seasonal Peaks:
# Strikes are likely higher during spring and fall, coinciding with bird migration seasons.
# Yearly Trends:
# If strikes have increased over time, it might reflect changes in aviation traffic or bird populations near airports.
# Implications:

# Implement targeted risk-reduction measures during peak migration months.
# Monitor and evaluate mitigation strategies for years with increasing trends.
# 3. Average Altitude of Bird Strikes by Top 5 Bird Species
# Insights:

# Larger migratory birds like geese or cranes may strike at higher average altitudes, compared to smaller birds like sparrows or pigeons.
# Each species may have distinct altitude ranges, which can guide species-specific risk management.
# Implications:

# Deploy species-specific countermeasures at critical altitudes where high-risk bird species are most active.
# Collaborate with wildlife agencies to manage high-risk species populations near airports.?
```
